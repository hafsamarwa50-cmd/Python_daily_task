```python
from abc import ABC, abstractmethod
import math

class Shape(ABC):

    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass


class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def perimeter(self):
        return 2 * math.pi * self.radius


class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)


c = Circle(5)
print("Circle Area:", c.area())
print("Circle Perimeter:", c.perimeter())

r = Rectangle(10, 5)
print("Rectangle Area:", r.area())
print("Rectangle Perimeter:", r.perimeter())
```

```python
from abc import ABC, abstractmethod
import math

class Shape3D(ABC):

    @abstractmethod
    def volume(self):
        pass

    @abstractmethod
    def surface_area(self):
        pass


class Cube(Shape3D):

    def __init__(self, side):
        self.side = side

    def volume(self):
        return self.side ** 3

    def surface_area(self):
        return 6 * self.side ** 2


class Sphere(Shape3D):

    def __init__(self, radius):
        self.radius = radius

    def volume(self):
        return (4/3) * math.pi * self.radius ** 3

    def surface_area(self):
        return 4 * math.pi * self.radius ** 2


c = Cube(4)
print("Cube Volume:", c.volume())
print("Cube Surface Area:", c.surface_area())

s = Sphere(5)
print("Sphere Volume:", s.volume())
print("Sphere Surface Area:", s.surface_area())
```

```python
from abc import ABC, abstractmethod

class BankAccount(ABC):
    total_accounts = 0

    def __init__(self, balance):
        self.balance = balance
        BankAccount.total_accounts += 1

    @abstractmethod
    def calculate_interest(self):
        pass

    @classmethod
    def show_total_accounts(cls):
        print("Total Accounts:", cls.total_accounts)


class SavingsAccount(BankAccount):
    MIN_BALANCE = 1000

    def __init__(self, balance):
        super().__init__(balance)

    def calculate_interest(self):
        return self.balance * 0.05

    def withdraw(self, amount):
        if self.balance - amount < SavingsAccount.MIN_BALANCE:
            print("Cannot withdraw. Minimum balance must be maintained.")
        else:
            self.balance -= amount
            print("Withdrawal Successful")
            print("Balance:", self.balance)


class CurrentAccount(BankAccount):
    OVERDRAFT_LIMIT = -5000

    def __init__(self, balance):
        super().__init__(balance)

    def calculate_interest(self):
        return 0

    def withdraw(self, amount):
        if self.balance - amount < CurrentAccount.OVERDRAFT_LIMIT:
            print("Overdraft limit exceeded.")
        else:
            self.balance -= amount
            print("Withdrawal Successful")
            print("Balance:", self.balance)


s1 = SavingsAccount(10000)
print("Savings Interest:", s1.calculate_interest())
s1.withdraw(9500)

c1 = CurrentAccount(2000)
c1.withdraw(6000)

BankAccount.show_total_accounts()
```

```python
from abc import ABC, abstractmethod
import math

class Shape(ABC):

    def __init__(self, color):
        self.color = color

    def display_color(self):
        print("Color:", self.color)

    @abstractmethod
    def area(self):
        pass

    @abstractmethod
    def perimeter(self):
        pass


class Rectangle(Shape):

    def __init__(self, color, length, width):
        super().__init__(color)
        self.length = length
        self.width = width

    def area(self):
        return self.length * self.width

    def perimeter(self):
        return 2 * (self.length + self.width)

    def display(self):
        super().display_color()
        print("Length:", self.length)
        print("Width:", self.width)
        print("Area:", self.area())
        print("Perimeter:", self.perimeter())


class Circle(Shape):

    def __init__(self, color, radius):
        super().__init__(color)
        self.radius = radius

    def area(self):
        return math.pi * self.radius ** 2

    def perimeter(self):
        return 2 * math.pi * self.radius

    def display(self):
        super().display_color()
        print("Radius:", self.radius)
        print("Area:", self.area())
        print("Perimeter:", self.perimeter())


r = Rectangle("Blue", 10, 5)
r.display()

print()

c = Circle("Red", 7)
c.display()
```

```python
from abc import ABC, abstractmethod

class Employee(ABC):

    def __init__(self, name, base_salary):
        self.name = name
        self.base_salary = base_salary

    def display_employee(self):
        print("Name:", self.name)
        print("Base Salary:", self.base_salary)

    @abstractmethod
    def calculate_bonus(self):
        pass


class Manager(Employee):

    def __init__(self, name, base_salary, extra_bonus):
        super().__init__(name, base_salary)
        self.extra_bonus = extra_bonus

    def calculate_bonus(self):
        return self.base_salary * 0.30 + self.extra_bonus

    def display_employee(self):
        super().display_employee()
        print("Extra Bonus:", self.extra_bonus)
        print("Total Bonus:", self.calculate_bonus())


class Developer(Employee):

    def __init__(self, name, base_salary):
        super().__init__(name, base_salary)

    def calculate_bonus(self):
        return self.base_salary * 0.20

    def display_employee(self):
        super().display_employee()
        print("Bonus:", self.calculate_bonus())


m = Manager("Rahul", 80000, 5000)
m.display_employee()

print()

d = Developer("Ali", 60000)
d.display_employee()
```

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):

    def __init__(self, brand, fuel_type):
        self.brand = brand
        self.fuel_type = fuel_type

    def display_vehicle(self):
        print("Brand:", self.brand)
        print("Fuel Type:", self.fuel_type)

    @abstractmethod
    def start(self):
        pass

    @abstractmethod
    def fuel_consumption(self, distance):
        pass


class ElectricCar(Vehicle):

    def __init__(self, brand, fuel_type):
        if fuel_type != "Electric":
            raise ValueError("Fuel type must be Electric")
        super().__init__(brand, fuel_type)

    def start(self):
        print("Electric Car starts silently.")

    def fuel_consumption(self, distance):
        print("Battery Used:", distance * 0.15, "kWh")

    def display_vehicle(self):
        super().display_vehicle()


class PetrolCar(Vehicle):

    def __init__(self, brand, fuel_type):
        if fuel_type != "Petrol":
            raise ValueError("Fuel type must be Petrol")
        super().__init__(brand, fuel_type)

    def start(self):
        print("Petrol Car starts with ignition.")

    def fuel_consumption(self, distance):
        print("Petrol Used:", distance / 15, "Litres")

    def display_vehicle(self):
        super().display_vehicle()


e = ElectricCar("Tesla", "Electric")
e.display_vehicle()
e.start()
e.fuel_consumption(120)

print()

p = PetrolCar("Toyota", "Petrol")
p.display_vehicle()
p.start()
p.fuel_consumption(150)
```

```python
from abc import ABC, abstractmethod

class Report(ABC):

    def __init__(self, title, data):
        self.title = title
        self.data = data

    def header(self):
        print("========== REPORT ==========")
        print("Title:", self.title)

    def footer(self):
        print("========== END OF REPORT ==========")

    @abstractmethod
    def generate(self):
        pass


class PDFReport(Report):

    def __init__(self, title, data):
        super().__init__(title, data)

    def generate(self):
        super().header()
        print("Generating PDF Report...")
        print("Data:", self.data)
        super().footer()


class HTMLReport(Report):

    def __init__(self, title, data):
        super().__init__(title, data)

    def generate(self):
        super().header()
        print("Generating HTML Report...")
        print("Data:", self.data)
        super().footer()


pdf = PDFReport("Sales Report", "Total Sales = ₹5,00,000")
pdf.generate()

print()

html = HTMLReport("Student Report", "Total Students = 120")
html.generate()
```
