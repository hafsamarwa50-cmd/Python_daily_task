```python
text = "Python Programming"

result = [ch for ch in text if ch.lower() not in "aeiou"]

print(result)
print("".join(result))
```

## 3. ADVANCED AND NESTED LIST COMPREHENSION:

```python
table = [[i * j for j in range(1, 6)] for i in range(1, 6)]

print(table)
```

```python
nested = [[1, 2], [3, 4], [5, 6]]

flat = [num for sublist in nested for num in sublist]

print(flat)
```

```python
words = ["Python", "Java"]

characters = [ch for word in words for ch in word]

print(characters)
```

```python
matrix = [[i * 3 + j + 1 for j in range(3)] for i in range(3)]

print(matrix)
```

```python
fruits = ["apple", "banana", "cherry"]

result = [(index, value) for index, value in enumerate(fruits)]

print(result)
```

```python
list1 = [1, 2, 3, 4, 5]
list2 = [3, 4, 5, 6, 7]

common = [x for x in list1 if x in list2]

print(common)
```

```python
import math

factorials = [math.factorial(i) for i in range(1, 7)]

print(factorials)
```

```python
print([[j for j in range(i + 1)] for i in range(4)])
```

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]

result = [(x, y) for x in list1 for y in list2 if x + y > 6]

print(result)
```

## HARD QUESTIONS ADVANCED LEVEL:

```python
result = [str(i) + " (multiple)" for i in range(1, 51) if i % 3 == 0 and i % 5 == 0]

print(result)
```

```python
data = ["10", "abc", "25.5", "30", "", "45.0", "xyz", "60"]
```

```python
data = ["10", "abc", "25.5", "30", "", "45.0", "xyz", "60"]

result = []

for x in data:
    try:
        result.append(int(float(x)))
    except ValueError:
        pass

print(result)
```

```python
names = ["Sumit", "Rahul", "Priya", "Aman", "Neha"]
scores = [85, 92, 78, 95, 88]
```

```python
result = [
    (index, name, "Excellent" if score >= 90 else "Good")
    for index, (name, score) in enumerate(zip(names, scores))
]

print(result)
```

```python
table = [
    [
        str(i * j) + (" (ok)" if (i * j) % 2 == 0 else " (odd)")
        for j in range(1, 6)
        if (i * j) % 7 != 0
    ]
    for i in range(1, 6)
]

print(table)
```

```python
result = []

for i in range(1, 21):
    if i % 2 == 0:
        result.append(str(pow(i, 2)) + " → " + str(abs(i - 10)))

print(result)
```

```python
[
'4 → 8',
'16 → 6',
'36 → 4',
'64 → 2',
'100 → 0',
'144 → 2',
'196 → 4',
'256 → 6',
'324 → 8',
'400 → 10'
]
```

```python
result = [f"{pow(i, 2)} → {abs(i - 10)}" for i in range(1, 21) if i % 2 == 0]

print(result)
```
