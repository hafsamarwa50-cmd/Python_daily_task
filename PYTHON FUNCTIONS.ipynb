# FUNCTIONS WITH DEFAULT PARAMETERS :

```python
def greet_user(name="Guest"):
    print("Hello", name)

greet_user()
greet_user("Hafsa")
```

```python
def power(base, exp=2):
    return base ** exp

print(power(5))
print(power(2, 3))
```

```python
def student_info(name, age=18, city="Delhi"):
    print("Name:", name)
    print("Age:", age)
    print("City:", city)

student_info("Ali")
student_info("Sara", 20, "Hyderabad")
```

```python
def add_numbers(a, b=0, c=0):
    return a + b + c

print(add_numbers(10))
print(add_numbers(10, 20))
print(add_numbers(10, 20, 30))
```

```python
def repeat_message(msg, times=3):
    for i in range(times):
        print(msg)

repeat_message("Hello")
repeat_message("Python", 5)
```

```python
def calculate_discount(price, discount=10):
    final_price = price - (price * discount / 100)
    return final_price

print(calculate_discount(1000))
print(calculate_discount(1000, 20))
```

```python
def introduce(name="Unknown", profession="Student"):
    print("Name:", name)
    print("Profession:", profession)

introduce()
introduce("Hafsa", "Developer")
```

# POSITIONAL ARGUMENTS & KEYWORD ARGUMENTS :

```python
def full_name(first, last):
    print(first, last)

full_name("Hafsa", "Marwa")
```

```python
def details(name, age, city):
    print(name, age, city)

# Positional
details("Ali", 22, "Mumbai")

# Keyword
details(name="Sara", age=20, city="Delhi")
```

```python
def calculate(a, b, operation):
    if operation == "+":
        print(a + b)
    elif operation == "-":
        print(a - b)
    elif operation == "*":
        print(a * b)
    elif operation == "/":
        print(a / b)

calculate(a=20, b=10, operation="+")
calculate(a=20, b=10, operation="*")
```

```python
def order(item, quantity, price):
    total = quantity * price
    print("Item:", item)
    print("Quantity:", quantity)
    print("Price:", price)
    print("Total:", total)

order("Book", quantity=2, price=300)
```

# ARGS (MULTIPLE POSITIONAL ARGUMENTS):

```python
def add_all(*args):
    return sum(args)

print(add_all(10, 20, 30))
print(add_all(5, 15, 25, 35))
```

```python
def print_all(*args):
    for value in args:
        print(value)

print_all("Python", 100, 25.5, True)
```

```python
def find_max(*args):
    return max(args)

print(find_max(10, 50, 20, 90, 40))
```

```python
def concatenate(*args):
    return " ".join(args)

print(concatenate("Hello", "Welcome", "to", "Python"))
```

```python
def multiply(*args):
    result = 1
    for i in args:
        result *= i
    return result

print(multiply(2, 3, 4))
print(multiply(5, 10))
```

```python
def average_marks(*args):
    return sum(args) / len(args)

print("Average:", average_marks(80, 90, 85, 95))
```

```python
def info(*args):
    for value in args:
        print("Value:", value, "Type:", type(value))

info(10, "Python", 25.5, True)
```

# KWARGS (MULTIPLE KEYWORD ARGUMENTS) :

```python
def student_info(**kwargs):
    for key, value in kwargs.items():
        print(key, ":", value)

student_info(name="Hafsa", age=20, city="Hyderabad")
```

```python
def create_profile(**kwargs):
    print("Profile Details")
    for key, value in kwargs.items():
        print(key, ":", value)

create_profile(
    name="Ali",
    age=22,
    city="Delhi",
    course="Python"
)
```

```python
def calculate_total(**kwargs):
    total = sum(kwargs.values())

    for item, price in kwargs.items():
        print(item, ":", price)

    print("Total Price:", total)

calculate_total(Book=300, Pen=50, Bag=1200)
```

```python
def display_info(name, **kwargs):
    print("Name:", name)

    for key, value in kwargs.items():
        print(key, ":", value)

display_info(
    "Hafsa",
    age=20,
    city="Hyderabad",
    profession="Student"
)
```

```python
def save_data(**kwargs):
    print("Saving data:")

    for key, value in kwargs.items():
        print(key, ":", value)

save_data(
    name="Sara",
    age=21,
    city="Mumbai",
    course="Data Science"
)
```

# HARD ON PYTHON FUNCTIONS :

```python
def process_data(*args, **kwargs):
    result = {}

    if kwargs.get("add"):
        result["add"] = sum(args)

    if kwargs.get("multiply"):
        product = 1
        for i in args:
            product *= i
        result["multiply"] = product

    if kwargs.get("power"):
        power_value = kwargs.get("power_value", 2)
        result["power"] = [num ** power_value for num in args]

    return result


print(process_data(2, 3, 4, add=True, multiply=True))
print(process_data(2, 3, 4, add=True, power=True, power_value=3))
```

```python
def make_multiplier(n):

    def multiplier(x):
        return x * n

    return multiplier


double = make_multiplier(2)
triple = make_multiplier(3)

print(double(10))
print(double(25))

print(triple(10))
print(triple(25))
```

```python
def smart_divide(func):

    def wrapper(a, b):

        if b == 0:
            print("Cannot divide by zero!")
            return None

        return func(a, b)

    return wrapper


@smart_divide
def divide(a, b):
    return a / b


print(divide(20, 5))
print(divide(10, 0))
```

```python
def flexible_calc(*args, operation="add", **kwargs):

    if len(args) == 0:
        return "No numbers provided."

    if operation == "add":
        result = sum(args)

    elif operation == "subtract":
        result = args[0]
        for i in args[1:]:
            result -= i

    elif operation == "multiply":
        result = 1
        for i in args:
            result *= i

    elif operation == "divide":
        result = args[0]

        for i in args[1:]:
            if i == 0:
                return "Division by zero is not allowed."
            result /= i

    else:
        return "Invalid operation."

    if kwargs.get("round_result"):
        result = round(result, 2)

    return result


print(flexible_calc(10, 20, 30))

print(flexible_calc(100, 20, 5, operation="divide"))

print(flexible_calc(5, 2, operation="divide", round_result=True))

print(flexible_calc(2, 3, 4, operation="multiply"))

print(flexible_calc(10, 20, 5, operation="subtract"))
```
