```python
class Temperature:
    def __init__(self, celsius):
        self.__celsius = celsius

    def get_celsius(self):
        return self.__celsius

    def set_celsius(self, celsius):
        self.__celsius = celsius

    def to_fahrenheit(self):
        return (self.__celsius * 9 / 5) + 32

temp = Temperature(25)

print("Celsius:", temp.get_celsius())

print("Fahrenheit:", temp.to_fahrenheit())

temp.set_celsius(30)

print("Updated Fahrenheit:", temp.to_fahrenheit())
```

# HARD QUESTIONS ON OOP:

```python
class Employee:
    total_employees = 0
    last_updated_salary = 0

    def __init__(self, name, salary):
        self.name = name
        self.__salary = salary
        Employee.total_employees += 1

    def get_salary(self):
        return self.__salary

    def set_salary(self, new_salary):
        if new_salary < 5000:
            print("Salary cannot be less than 5000")
        elif new_salary > self.__salary * 1.20:
            print("Salary increase cannot exceed 20%")
        else:
            self.__salary = new_salary
            Employee.last_updated_salary = new_salary
            print("Salary Updated")

    @classmethod
    def get_total_employees(cls):
        return cls.total_employees


emp1 = Employee("Ali", 10000)
emp2 = Employee("Sara", 15000)

print("Salary:", emp1.get_salary())

emp1.set_salary(11500)

print("Updated Salary:", emp1.get_salary())

print("Last Updated Salary:", Employee.last_updated_salary)

print("Total Employees:", Employee.get_total_employees())
```

```python
class BankAccount:
    total_accounts = 0
    history = []

    def __init__(self, balance):
        self.__balance = balance
        BankAccount.total_accounts += 1

    def get_balance(self):
        return self.__balance

    def set_balance(self, balance):
        if balance < 1000:
            print("Minimum balance is 1000")
            return

        if balance > 100000:
            balance = balance + (balance * 0.05)

        self.__balance = balance
        BankAccount.history.append(balance)
        print("Balance Updated")

    @classmethod
    def show_total_accounts(cls):
        return cls.total_accounts


acc1 = BankAccount(5000)

acc1.set_balance(120000)

print("Balance:", acc1.get_balance())

print("History:", BankAccount.history)

print("Total Accounts:", BankAccount.show_total_accounts())
```

```python
class Student:
    total_students = 0

    def __init__(self, marks):
        self.__marks = marks
        self.status = "Passed"
        Student.total_students += 1

    def get_marks(self):
        return self.__marks

    def set_marks(self, marks):

        if len(marks) != 5:
            print("Enter marks for 5 subjects")
            return

        for i in marks:
            if i < 0 or i > 100:
                print("Invalid Marks")
                return

        old_avg = sum(self.__marks) / 5
        new_avg = sum(marks) / 5

        if new_avg < old_avg:
            print("Average cannot decrease")
            return

        self.__marks = marks

        if min(marks) < 40:
            self.status = "Failed"
        else:
            self.status = "Passed"


s1 = Student([60, 70, 80, 75, 65])

s1.set_marks([80, 85, 90, 88, 92])

print("Marks:", s1.get_marks())

print("Status:", s1.status)

print("Total Students:", Student.total_students)
```

```python
class Product:
    total_revenue = 0

    def __init__(self, price):
        self.__price = price
        self.original_price = price
        self.__discount = 0

    def get_price(self):
        return self.__price

    def get_discount(self):
        return self.__discount

    def set_price(self, price):

        if price < self.original_price * 0.80:
            print("Price cannot be less than 80% of original price")
            return

        self.__price = price

        if price > 5000:
            self.__discount = 15

        Product.total_revenue += price

        print("Price Updated")


p1 = Product(4000)

p1.set_price(6000)

print("Price:", p1.get_price())

print("Discount:", p1.get_discount(), "%")

print("Total Revenue:", Product.total_revenue)
```

```python
class User:
    senior_citizen = 0

    def __init__(self, password, age):
        self.__password = password
        self.__age = age
        self.is_senior = False

        if age >= 60:
            self.is_senior = True
            User.senior_citizen += 1

    def get_password(self):
        return self.__password

    def get_age(self):
        return self.__age

    def set_password(self, password):

        if len(password) < 8:
            print("Password must have at least 8 characters")
            return

        digit = False

        for ch in password:
            if ch.isdigit():
                digit = True

        if digit == False:
            print("Password must contain at least one digit")
            return

        self.__password = password

    def set_age(self, age):

        if age < self.__age:
            print("Age cannot be decreased")
            return

        self.__age = age

        if age >= 60 and self.is_senior == False:
            self.is_senior = True
            User.senior_citizen += 1

    def update(self, password, age):
        self.set_password(password)
        self.set_age(age)


u1 = User("SUMIT12345", 55)

u1.update("Marwa123", 62)

print("Password:", u1.get_password())

print("Age:", u1.get_age())

print("Senior Citizen:", u1.is_senior)

print("Total Senior Citizens:", User.senior_citizen)
```
