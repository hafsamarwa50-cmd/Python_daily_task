# Interview Challenge:

```python
from functools import reduce

students = [
    {"name": "Ram", "marks": [70, 80, 90]},
    {"name": "Amit", "marks": [90, 92, 94]},
    {"name": "Neha", "marks": [60, 65, 70]},
    {"name": "Raj", "marks": [88, 84, 90]}
]

# 1. Calculate average marks
averages = list(map(
    lambda s: {
        "name": s["name"],
        "average": reduce(lambda x, y: x + y, s["marks"]) / len(s["marks"])
    },
    students
))

print("Average Marks:")
print(averages)

# 2. Sort students by average (highest first)
sorted_students = sorted(
    averages,
    key=lambda x: x["average"],
    reverse=True
)

print("\nSorted by Average:")
print(sorted_students)

# 3. Find the topper
topper = max(
    averages,
    key=lambda x: x["average"]
)

print("\nTopper:")
print(topper)

# 4. Filter students with average > 80
filtered = list(filter(
    lambda x: x["average"] > 80,
    averages
))

print("\nAverage > 80:")
print(filtered)

# 5. Print only names
names = list(map(
    lambda x: x["name"],
    filtered
))

print("\nStudent Names:")
print(names)
```

# INTERVIEW QUESTIONS OPTIONAL:

```python
from functools import reduce

students = [
    {"name":"Ram","marks":{"Math":80,"Python":95,"SQL":88}},
    {"name":"Amit","marks":{"Math":92,"Python":90,"SQL":91}},
    {"name":"Neha","marks":{"Math":85,"Python":99,"SQL":94}},
    {"name":"Raj","marks":{"Math":78,"Python":70,"SQL":80}}
]

# 1. Calculate average
averages = list(map(
    lambda s: {
        "name": s["name"],
        "average": reduce(lambda x, y: x + y, s["marks"].values()) / len(s["marks"])
    },
    students
))

print("Averages:", averages)

# 2. Sort by average
sorted_students = sorted(
    averages,
    key=lambda x: x["average"],
    reverse=True
)

print("Sorted:", sorted_students)

# 3. Topper
print("Topper:", max(averages, key=lambda x: x["average"]))

# 4. Lowest scorer
print("Lowest:", min(averages, key=lambda x: x["average"]))

# 5. Average > 90
above90 = list(filter(lambda x: x["average"] > 90, averages))
print("Above 90:", above90)

# 6. Names only
print("Names:", list(map(lambda x: x["name"], averages)))

# 7. Total marks of all students
total_marks = reduce(
    lambda x, y: x + y,
    map(lambda s: sum(s["marks"].values()), students)
)

print("Total Marks:", total_marks)
```

```python
from functools import reduce

employees = [
    {"id":101,"name":"Ram","dept":"IT","salary":55000},
    {"id":102,"name":"Amit","dept":"HR","salary":40000},
    {"id":103,"name":"Neha","dept":"IT","salary":70000},
    {"id":104,"name":"Raj","dept":"Sales","salary":45000},
    {"id":105,"name":"Anu","dept":"IT","salary":85000},
    {"id":106,"name":"Kiran","dept":"Sales","salary":62000}
]

print("Highest:", max(employees, key=lambda x: x["salary"]))

print("Lowest:", min(employees, key=lambda x: x["salary"]))

average = reduce(
    lambda x, y: x + y,
    map(lambda x: x["salary"], employees)
) / len(employees)

print("Average Salary:", average)

print("Above Average:",
      list(filter(lambda x: x["salary"] > average, employees)))

print("Sorted:",
      sorted(employees,
             key=lambda x: (x["dept"], -x["salary"])))

it_salary = reduce(
    lambda x, y: x + y,
    map(lambda x: x["salary"],
        filter(lambda x: x["dept"] == "IT", employees))
)

print("Total IT Salary:", it_salary)

print("Names:",
      list(map(lambda x: x["name"], employees)))

second = sorted(
    employees,
    key=lambda x: x["salary"],
    reverse=True
)[1]

print("Second Highest:", second)
```

```python
from functools import reduce

products = [
    {"name":"Laptop","price":65000,"rating":4.5},
    {"name":"Mouse","price":800,"rating":4.1},
    {"name":"Keyboard","price":1800,"rating":4.8},
    {"name":"Monitor","price":14000,"rating":4.6},
    {"name":"Tablet","price":28000,"rating":4.3},
    {"name":"Phone","price":42000,"rating":4.9}
]

print(max(products, key=lambda x: x["price"]))

print(min(products, key=lambda x: x["price"]))

print(list(filter(lambda x: x["price"] > 20000, products)))

print(list(filter(lambda x: x["rating"] > 4.5, products)))

print(sorted(products,
             key=lambda x: (-x["rating"], x["price"])))

average = reduce(
    lambda x, y: x + y,
    map(lambda x: x["price"], products)
) / len(products)

print("Average Price:", average)

inventory = reduce(
    lambda x, y: x + y,
    map(lambda x: x["price"], products)
)

print("Inventory:", inventory)

names = sorted(
    map(lambda x: x["name"], products)
)

print(names)
```

```python
from functools import reduce

orders = [
    {"id":1,"customer":"Ram","amount":4500,"status":"Delivered"},
    {"id":2,"customer":"Amit","amount":7000,"status":"Pending"},
    {"id":3,"customer":"Raj","amount":2000,"status":"Delivered"},
    {"id":4,"customer":"Neha","amount":9000,"status":"Delivered"},
    {"id":5,"customer":"Anu","amount":3500,"status":"Cancelled"},
    {"id":6,"customer":"Ravi","amount":10000,"status":"Delivered"}
]

delivered = list(filter(lambda x: x["status"] == "Delivered", orders))

total = reduce(
    lambda x, y: x + y,
    map(lambda x: x["amount"], delivered)
)

print("Total Delivered:", total)

print("Average:", total / len(delivered))

print("Largest:", max(delivered, key=lambda x: x["amount"]))

print("Pending:",
      list(filter(lambda x: x["status"] == "Pending", orders)))

print(sorted(orders,
             key=lambda x: x["amount"],
             reverse=True))

print(list(map(lambda x: x["customer"], delivered)))

print("Count:", len(delivered))

print("Maximum Customer:",
      max(delivered, key=lambda x: x["amount"]))
```

```python
from functools import reduce

students = [
    {"name":"Ram","age":22,"marks":[80,70,90],"skills":["Python","SQL"]},
    {"name":"Amit","age":20,"marks":[95,92,88],"skills":["Python","Power BI","SQL"]},
    {"name":"Neha","age":24,"marks":[85,90,80],"skills":["Excel","Python"]},
    {"name":"Raj","age":21,"marks":[65,70,72],"skills":["SQL"]}
]

averages = list(map(
    lambda s: {
        "name": s["name"],
        "age": s["age"],
        "skills": s["skills"],
        "average": sum(s["marks"]) / len(s["marks"])
    },
    students
))

overall = reduce(
    lambda x, y: x + y,
    map(lambda x: x["average"], averages)
) / len(averages)

result = {
    "Topper":
        max(averages, key=lambda x: x["average"])["name"],

    "Average":
        overall,

    "Python Students":
        list(map(lambda x: x["name"],
        filter(lambda x: "Python" in x["skills"], averages))),

    "SQL Students":
        list(map(lambda x: x["name"],
        filter(lambda x: "SQL" in x["skills"], averages))),

    "Above Average":
        list(map(lambda x: x["name"],
        filter(lambda x: x["average"] > overall, averages))),

    "Total Students":
        len(averages)
}

print(result)

print("Lowest:",
      min(averages, key=lambda x: x["average"]))

print("Sorted:",
      sorted(averages,
             key=lambda x: x["average"],
             reverse=True))

print("Python:",
      list(map(lambda x: x["name"],
      filter(lambda x: "Python" in x["skills"], averages))))

print("Python & SQL:",
      list(map(lambda x: x["name"],
      filter(lambda x:
      all(skill in x["skills"] for skill in ["Python", "SQL"]),
      averages))))

print("Age >21:",
      list(map(lambda x: x["name"],
      filter(lambda x: x["age"] > 21, averages))))

print("Most Skills:",
      max(averages, key=lambda x: len(x["skills"])))

print("Alphabetical:",
      sorted(map(lambda x: x["name"], averages)))

print("Python Count:",
      len(list(filter(lambda x: "Python" in x["skills"], averages))))

print("SQL Count:",
      len(list(filter(lambda x: "SQL" in x["skills"], averages))))

print("Names:",
      list(map(lambda x: x["name"], averages)))
```
