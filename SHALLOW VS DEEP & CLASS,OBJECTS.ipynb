```python
import copy

original = [1, 2, [3, 4], 5]

shallow = copy.copy(original)

shallow[2][0] = 99

print("Original List:", original)
print("Shallow Copy :", shallow)
```

```python
import copy

original = [1, 2, [3, 4], 5]

deep = copy.deepcopy(original)

deep[2][0] = 99

print("Original List:", original)
print("Deep Copy    :", deep)
```

```python
def manual_deep_copy(lst):

    new_list = []

    for item in lst:
        if isinstance(item, list):
            new_list.append(manual_deep_copy(item))
        else:
            new_list.append(item)

    return new_list


original = [1, 2, [3, 4], 5]

copied = manual_deep_copy(original)

copied[2][0] = 99

print("Original:", original)
print("Copied  :", copied)
```

```python
import copy

data = [10, [20, 30], {"name": "Sumit"}, 40]

shallow = copy.copy(data)

deep = copy.deepcopy(data)

data[1][0] = 99
data[2]["name"] = "Raju"

print("Original:", data)
print("Shallow :", shallow)
print("Deep    :", deep)
```

```python
import copy

def check_copy(copy_list, original):

    copy_list[1][0] = 99

    if original[1][0] == 99:
        print("It is a Shallow Copy")
    else:
        print("It is a Deep Copy")


original = [1, [2, 3], 4]

shallow = copy.copy(original)

deep = copy.deepcopy(original)

check_copy(shallow, original)

original = [1, [2, 3], 4]

check_copy(deep, original)
```

```python
import copy

class Student:

    def __init__(self, name, marks):
        self.name = name
        self.marks = marks


s1 = Student("Hafsa", [80, 85, 90])

shallow = copy.copy(s1)

deep = copy.deepcopy(s1)

shallow.marks[0] = 100

print("After Shallow Copy")
print("Original Marks:", s1.marks)
print("Shallow Marks :", shallow.marks)

deep.marks[1] = 95

print("\nAfter Deep Copy")
print("Original Marks:", s1.marks)
print("Deep Marks    :", deep.marks)
```

```python
class Student:

    def __init__(self, name, age, marks):
        self.name = name
        self.age = age
        self.marks = marks

    def display(self):
        print("Name :", self.name)
        print("Age :", self.age)
        print("Marks :", self.marks)


student1 = Student("Hafsa", 20, 92)
student2 = Student("Marwa", 21, 88)

print("Student 1")
student1.display()

print("\nStudent 2")
student2.display()
```

```python
class Employee:

    def __init__(self, emp_id, name, salary):
        self.emp_id = emp_id
        self.name = name
        self.salary = salary

    def increase_salary(self):
        self.salary = self.salary + 5000
        print("Updated Salary :", self.salary)


emp = Employee(101, "Rahul", 30000)

print("Employee Name :", emp.name)

emp.increase_salary()
```

```python
class Mobile:

    def __init__(self, brand, model, price):
        self.brand = brand
        self.model = model
        self.price = price

    def discount(self):
        self.price = self.price - (self.price * 10 / 100)
        print("Price after Discount :", self.price)


mobile = Mobile("Samsung", "Galaxy A55", 40000)

print("Original Price :", mobile.price)

mobile.discount()
```

```python
class Laptop:

    def __init__(self, brand, ram, storage):
        self.brand = brand
        self.ram = ram
        self.storage = storage

    def upgrade_ram(self):
        self.ram = self.ram + 8
        print("Updated RAM :", self.ram, "GB")


laptop = Laptop("HP", 8, 512)

print("Current RAM :", laptop.ram, "GB")

laptop.upgrade_ram()
```

```python
class BankAccount:

    def __init__(self, account_holder, balance):
        self.account_holder = account_holder
        self.balance = balance

    def deposit(self, amount):
        self.balance = self.balance + amount
        print("Deposited:", amount)
        print("Balance:", self.balance)

    def withdraw(self, amount):
        if amount <= self.balance:
            self.balance = self.balance - amount
            print("Withdrawn:", amount)
            print("Balance:", self.balance)
        else:
            print("Insufficient Balance")


account = BankAccount("Hafsa", 10000)

account.deposit(5000)

account.withdraw(3000)

account.withdraw(15000)
```

```python
class Car:

    def __init__(self, brand, speed):
        self.brand = brand
        self.speed = speed

    def increase_speed(self, value):
        if self.speed + value <= 200:
            self.speed = self.speed + value
            print("Current Speed:", self.speed, "km/h")
        else:
            print("Maximum speed is 200 km/h")


car = Car("Toyota", 150)

car.increase_speed(30)

car.increase_speed(40)
```

```python
class Temperature:

    def __init__(self, temperature):
        self.temperature = temperature

    def set_temperature(self, value):
        if value >= -20 and value <= 50:
            self.temperature = value
            print("Temperature:", self.temperature, "°C")
        else:
            print("Invalid Temperature")


temp = Temperature(25)

temp.set_temperature(40)

temp.set_temperature(60)
```

```python
class Movie:

    def __init__(self, movie_name, seats):
        self.movie_name = movie_name
        self.seats = seats

    def book_ticket(self, number):
        if number <= self.seats:
            self.seats = self.seats - number
            print("Booking Successful")
            print("Seats Left:", self.seats)
        else:
            print("Seats Not Available")


movie = Movie("3 Idiots", 50)

movie.book_ticket(5)

movie.book_ticket(60)
```

```python
class Library:

    def __init__(self, book_name, available_copies):
        self.book_name = book_name
        self.available_copies = available_copies

    def borrow_book(self):
        if self.available_copies > 0:
            self.available_copies = self.available_copies - 1
            print("Book Borrowed")
            print("Available Copies:", self.available_copies)
        else:
            print("Book Not Available")


library = Library("Python Programming", 2)

library.borrow_book()

library.borrow_book()

library.borrow_book()
```

```python
class Product:

    def __init__(self, product_name, quantity):
        self.product_name = product_name
        self.quantity = quantity

    def sell(self, qty):
        if qty <= self.quantity:
            self.quantity = self.quantity - qty
            print("Product Sold")
            print("Remaining Quantity:", self.quantity)
        else:
            print("Not Enough Stock")


product = Product("Laptop", 20)

product.sell(5)

product.sell(18)
```

```python
class ATM:

    def __init__(self, balance):
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print("Balance:", self.balance)
        else:
            print("Invalid Deposit")

    def withdraw(self, amount):
        if amount <= 0:
            print("Invalid Amount")
        elif amount > self.balance:
            print("Insufficient Balance")
        else:
            self.balance -= amount
            print("Balance:", self.balance)


atm = ATM(10000)

atm.deposit(2000)

atm.withdraw(3000)

atm.withdraw(0)

atm.withdraw(15000)
```

```python
class Student:

    def __init__(self, total_classes, attended_classes):
        self.total_classes = total_classes
        self.attended_classes = attended_classes

    def mark_attendance(self, number):
        if self.attended_classes + number <= self.total_classes:
            self.attended_classes += number

            percentage = (self.attended_classes / self.total_classes) * 100

            print("Attendance:", percentage, "%")
        else:
            print("Attendance exceeds total classes")


student = Student(100, 70)

student.mark_attendance(10)

student.mark_attendance(30)
```

```python
class Bus:

    def __init__(self, bus_name, total_seats, booked_seats):
        self.bus_name = bus_name
        self.total_seats = total_seats
        self.booked_seats = booked_seats

    def reserve(self, number):
        if self.booked_seats + number <= self.total_seats:
            self.booked_seats += number

            print("Reservation Successful")

            print("Remaining Seats:", self.total_seats - self.booked_seats)
        else:
            print("Seats Not Available")


bus = Bus("Express", 50, 30)

bus.reserve(10)

bus.reserve(15)
```

```python
class ShoppingCart:

    def __init__(self, customer_name, wallet_balance):
        self.customer_name = customer_name
        self.wallet_balance = wallet_balance
        self.cart_total = 0

    def add_item(self, price):
        self.cart_total += price
        print("Cart Total:", self.cart_total)

    def remove_item(self, price):
        if self.cart_total - price >= 0:
            self.cart_total -= price
            print("Cart Total:", self.cart_total)
        else:
            print("Cart total cannot become negative")

    def checkout(self):
        if self.wallet_balance >= self.cart_total:
            self.wallet_balance -= self.cart_total

            print("Payment Successful")

            print("Wallet Balance:", self.wallet_balance)

            self.cart_total = 0
        else:
            print("Insufficient Wallet Balance")


cart = ShoppingCart("Hafsa", 5000)

cart.add_item(1200)

cart.add_item(800)

cart.remove_item(500)

cart.checkout()
```

# BONUS CHALLENGE:

```python
class Student:

    def __init__(self, student_id, name, age, course, total_fee, marks, attendance, total_classes):
        self.student_id = student_id
        self.name = name
        self.age = age
        self.course = course
        self.total_fee = total_fee
        self.paid_fee = 0
        self.marks = marks
        self.attendance = attendance
        self.total_classes = total_classes

    def update_marks(self, marks):
        if 0 <= marks <= 100:
            self.marks = marks
        else:
            print("Invalid Marks")

    def pay_fee(self, amount):
        if self.paid_fee + amount <= self.total_fee:
            self.paid_fee += amount
        else:
            print("Payment Exceeds Remaining Fee")

    def update_attendance(self, classes):
        if self.attendance + classes <= self.total_classes:
            self.attendance += classes
        else:
            print("Attendance Exceeds Total Classes")

    def change_course(self, course):
        self.course = course

    def attendance_percentage(self):
        return (self.attendance / self.total_classes) * 100

    def check_result(self):
        if self.marks >= 35:
            return "Pass"
        return "Fail"

    def scholarship(self):
        if self.marks > 90 and self.attendance_percentage() > 95 and self.paid_fee == self.total_fee:
            return "Yes"
        return "No"

    def display_report(self):
        print("\n------ STUDENT REPORT CARD ------")
        print("Student ID :", self.student_id)
        print("Name :", self.name)
        print("Age :", self.age)
        print("Course :", self.course)
        print("Marks :", self.marks)
        print("Result :", self.check_result())
        print("Attendance % :", self.attendance_percentage())
        print("Fee Paid :", self.paid_fee)
        print("Remaining Fee :", self.total_fee - self.paid_fee)
        print("Scholarship Eligible :", self.scholarship())


student = Student(101, "Hafsa", 20, "IT", 100000, 88, 90, 100)

student.update_marks(95)

student.pay_fee(50000)

student.update_attendance(8)

student.change_course("B.tech IT")

student.display_report()
```
